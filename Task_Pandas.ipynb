##Filter Dataframe

```python
# Filter harga lebih dari 20000 dan tahun 2017
filtered_multiple = df[(df['price'] > 20000) & (df['year'] == 2017)]
print('Filter harga lebih dari 20000 dan tahun 2017:')
print(filtered_multiple)
```

```python
# Filter tahun 2017 dan 2018
filtered_isin = df[df['year'].isin([2017, 2018])]
print('Filter tahun 2017 dan 2018:')
print(filtered_isin)
```

```python
# Filter dengan str accessor (misal: tipe bahan bakar 'Petrol')
filtered_str = df[df['fuelType'].str.contains('Petrol')]
print('Filter dengan bahan bakar Petrol:')
print(filtered_str)
```

```python
# Dapatkan 5 harga termahal
top_prices = df.nlargest(5, 'price')
print('5 harga termahal:')
print(top_prices)
```

```python
# Dapatkan 5 harga termurah
bottom_prices = df.nsmallest(5, 'price')
print('5 harga termurah:')
print(bottom_prices)
```

```python
# Menggunakan query (misal: tahun 2017 dan harga lebih dari 15000)
filtered_query = df.query('year == 2017 & price > 15000')
print("Filtered Query:\n", filtered_query)
```

##Group dataframe

```python
# Contoh pengelompokan berdasarkan tahun dan menghitung rata-rata harga
grouped_df = df.groupby('year')['price'].mean().reset_index()
print(grouped_df)
```

##Replace Dataframe

```python
# Mengganti 'Petrol' dengan 'Gasoline'
df['fuelType'] = df['fuelType'].replace('Petrol', 'Gasoline')

# Menampilkan hasilnya
df.head()
```

```python
# Mengganti 'Manual' menjadi 'M' dan 'Automatic' menjadi 'A'
df['transmission'] = df['transmission'].replace({'Manual': 'M', 'Automatic': 'A'})

# Menampilkan hasilnya
df.head()
```

```python
# Simpan DataFrame yang telah dimodifikasi ke dalam file CSV
df.to_csv('/content/drive/My Drive/dataset_modified.csv', index=False)

print("Dataset telah berhasil disimpan.")
```
